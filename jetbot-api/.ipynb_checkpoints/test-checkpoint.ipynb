```python
# Test JUST the camera (no asyncio)
from jetbot import Camera, bgr8_to_jpeg
import base64
import time

print('🧪 Testing JetBot camera in notebook...')

try:
    # Initialize camera with defaults (like working notebook example)
    camera = Camera.instance()
    print('✅ Camera.instance() successful!')
    
    # Give it time to initialize 
    time.sleep(2)
    
    # Test frame capture
    frame = camera.value
    if frame is not None:
        print(f'✅ Frame captured! Shape: {frame.shape}')
        
        # Test JPEG conversion
        jpeg_bytes = bgr8_to_jpeg(frame)
        print(f'✅ bgr8_to_jpeg successful! Size: {len(jpeg_bytes)} bytes')
        
        # Test base64 encoding
        base64_str = base64.b64encode(jpeg_bytes).decode('utf-8')
        print(f'✅ Base64 encoding successful! Length: {len(base64_str)} chars')
        
        print('🎉 Camera pipeline working in notebook!')
        
    else:
        print('⚠️ No frame captured yet')
        
except Exception as e:
    print(f'❌ Error: {e}')
    import traceback
    traceback.print_exc()
```

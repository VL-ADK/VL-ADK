```python
# Test JUST the camera (no asyncio)
from jetbot import Camera, bgr8_to_jpeg
import base64
import time

print('üß™ Testing JetBot camera in notebook...')

try:
    # Initialize camera with defaults (like working notebook example)
    camera = Camera.instance()
    print('‚úÖ Camera.instance() successful!')
    
    # Give it time to initialize 
    time.sleep(2)
    
    # Test frame capture
    frame = camera.value
    if frame is not None:
        print(f'‚úÖ Frame captured! Shape: {frame.shape}')
        
        # Test JPEG conversion
        jpeg_bytes = bgr8_to_jpeg(frame)
        print(f'‚úÖ bgr8_to_jpeg successful! Size: {len(jpeg_bytes)} bytes')
        
        # Test base64 encoding
        base64_str = base64.b64encode(jpeg_bytes).decode('utf-8')
        print(f'‚úÖ Base64 encoding successful! Length: {len(base64_str)} chars')
        
        print('üéâ Camera pipeline working in notebook!')
        
    else:
        print('‚ö†Ô∏è No frame captured yet')
        
except Exception as e:
    print(f'‚ùå Error: {e}')
    import traceback
    traceback.print_exc()
```
